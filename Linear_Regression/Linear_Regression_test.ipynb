### Testing Linear Regression algorithm made from scratch

```python
#importing necessary packages
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#creating data set
X, y = datasets.make_regression(n_samples=100, n_features=1, noise=20,random_state=4)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
#fitting the linear regressor and predicting the test ys 
from Linear_Regression import LinearRegression

regressor = LinearRegression(lr = .1)
regressor.fit(X_train, y_train)
predicted = regressor.predict(X_test)
```

```python
#finding the MSE and looking at how well the prediction line matches up with the actual values
def mse(y_true, y_predicted):
    return np.mean((y_true-y_predicted)**2)

test_mse_value = mse(y_test,predicted)
print(f"Test MSE: {test_mse_value}")

y_pred_line = regressor.predict(X)
plt.scatter(X_train, y_train, color="blue")
plt.scatter(X_test, y_test, color="green")
plt.plot(X, y_pred_line, color='black', label='Prediction')
```
